```python
csv_path = './itsleanin808.csv'
```

```python
df = pl.read_csv(
    csv_path,
    has_header=False,
    new_columns=['Artist', 'Album', 'Track', 'Timestamp']
)
```

```python
df
```

```python
unique_songs = df.select(["Artist", "Track"]).unique()
```

```python
unique_songs
```

```python
id_results = []
results = None
for artist, track in tqdm(unique_songs.iter_rows(), total=unique_songs.height):
    query = f"artist:{artist} track:{track}"
    spotify_id = None
    try:
        results = sp.search(q=query, type='track', limit=1)
        items = results['tracks']['items']
        if items:
            spotify_id = items[0]['id']
    except Exception as e:
        # Em produção, você pode querer logar o erro
        pass

    # Guardamos mesmo se for None, para saber que já buscamos
    id_results.append({
        "Artist": artist,
        "Track": track,
        "spotify_id": spotify_id
    })

    # Rate Limiting manual simples (opcional, o spotipy gerencia retries básicos)
    time.sleep(0.1)
```

```python
ids_df = pl.DataFrame(id_results)
```

```python
found_ids_df = ids_df.filter(pl.col("spotify_id").is_not_null())
id_list = found_ids_df.select("spotify_id").to_series().to_list()
```

```python
import socket
import requests

hostname = "api.spotify.com"

print("--- DIAGNÓSTICO DE REDE ---")
try:
    # 1. Teste de DNS: Para qual IP o Spotify está apontando?
    ip_address = socket.gethostbyname(hostname)
    print(f"1. DNS: '{hostname}' está resolvendo para o IP: {ip_address}")

    # Se o IP começar com 10., 192. ou 172., é um bloqueio local/rede interna.
    # O IP correto do Spotify geralmente muda, mas é um IP público (ex: 35.x.x.x).

    # 2. Teste de Conexão Direta (sem Spotipy)
    print("2. Tentando conexão HTTP direta...")
    response = requests.get("https://api.spotify.com", timeout=5)
    print(f"   Status Code: {response.status_code}")
    print(f"   URL Final: {response.url}")

except Exception as e:
    print(f"ERRO FATAL: {e}")
```
